```python
#Bootstrapping
import numpy as np
n=100;
S=np.random.beta(2,2,n)
def boostrap(S,n):
    m=S.size
    indx=np.random.randint(n,size=m)
    r=S[indx]
    return r;
m=50;
k=1000;
medians=np.zeros(k)
for i in range(k):
    medians[i]=np.median(boostrap(S,m))
print(medians.size)
print(np.mean(medians))
print(np.median(medians))
```

```python
sns.set()
ax=sns.distplot(medians)
```

```python
m=50;
k=1000;
tenper=np.zeros(k)
for i in range(k):
    tenper[i]=np.percentile(boostrap(S,m),10)
print(tenper.size)
print(np.mean(tenper))
print(np.median(tenper))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
n=100;
np.random.seed(100)
pm=np.random.pareto(a=5.0,size=n)
pw=np.random.pareto(a=4.0,size=n)
obs=abs(np.percentile(pw,50)-np.percentile(pm,50))
print(obs)

```

```python
sns.distplot(pw)
sns.distplot(pm)
```

```python

tk=abs(np.median(pw)-np.median(pm))
P=np.concatenate((pw,pm))
k=1000;
teststats=np.zeros(k)
for i in range(k):
    np.random.shuffle(P)
    temp_men,temp_women=P[:n],P[n:]
    teststats[i]=abs(np.percentile(temp_men,50)-np.percentile(temp_women,50))
kwargs={'cumulative':True}
sns.distplot(teststats,hist_kws=kwargs,kde_kws=kwargs)
plt.grid()
plt.show()
        
```

### the final step

```python
#now to see whether the null hypothesis is accepted or rejected
obs=abs(np.percentile(pw,50)-np.percentile(pm,50))
num_ho=np.sum(teststats>=obs)
prob=num_ho/k
print(prob)
```

```python
#KS test
from scipy import stats
n=100;
np.random.seed(100)
pm=np.random.pareto(a=5.0,size=n)
pw=np.random.pareto(a=4.0,size=n)
D,p=stats.ks_2samp(pw,pm)
print(D,p)
```
